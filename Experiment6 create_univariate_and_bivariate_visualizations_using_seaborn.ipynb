```python
import seaborn as sns
import pandas as pd

# Load the "tips" dataset
tips = sns.load_dataset("tips")
print(tips.head())


```

```python
plt.figure(figsize=(8,5))
sns.histplot(tips['total_bill'], kde=True, bins=20, color='skyblue')
plt.title("Distribution of Total Bill")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='day', data=tips, palette='Set2')
plt.title("Count of Days")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y='tip', data=tips, color='lightgreen')
plt.title("Boxplot of Tips")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', palette='cool')
plt.title("Scatterplot of Total Bill vs Tip (by Gender)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='tip', data=tips, palette='Set3')
plt.title("Tips Distribution across Days")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.violinplot(x='sex', y='tip', data=tips, palette='muted')
plt.title("Violin Plot of Tips by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x='total_bill', y='tip', data=tips, scatter_kws={'alpha':0.6})
plt.title("Regression Plot of Total Bill vs Tip")
plt.show()
```
